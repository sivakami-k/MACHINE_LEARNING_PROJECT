```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
outcome_enc=lbl.fit_transform(df['Heart Disease'])

df['Heart Disease']=outcome_enc
df
```

```python
print(df.corr()["Heart Disease"].abs().sort_values(ascending=False))
```

```python
sc= MinMaxScaler(feature_range=(0,1))
data_scaled= sc.fit_transform(df)
```

```python
data_scaled= pd.DataFrame(data_scaled)
data_scaled
```

```python
x = data_scaled.iloc[:,:-1]
print(x)
y = data_scaled.iloc[:,-1].values
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=42)
```

KNN ALGORITHM


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model1=KNeighborsClassifier()
model1.fit(xtrain,ytrain)
y_pred=model1.predict(xtest)
```

```python
acc1=accuracy_score(ytest,y_pred)
print(acc1)
```

```python
cm1=confusion_matrix(ytest,y_pred)
print(cm1)
dis1=ConfusionMatrixDisplay(confusion_matrix=cm1)
dis1.plot()
plt.show()

```

Naive Bayes


```python
from sklearn.naive_bayes import GaussianNB
```

```python
model2=GaussianNB()
model2.fit(xtrain,ytrain)
y_pred=model2.predict(xtest)
```

```python
acc2=accuracy_score(ytest,y_pred)
print(acc2)
```

```python
cm2=confusion_matrix(ytest,y_pred)
print(cm2)
dis2=ConfusionMatrixDisplay(confusion_matrix=cm2)
dis2.plot()
plt.show()
```

Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model3=DecisionTreeClassifier()
model3.fit(xtrain,ytrain)
y_pred=model3.predict(xtest)
```

```python
acc3=accuracy_score(ytest,y_pred)
print(acc3)
```

```python
cm3=confusion_matrix(ytest,y_pred)
print(cm3)
dis3=ConfusionMatrixDisplay(confusion_matrix=cm3)
dis3.plot()
plt.show()
```

RANDOM FOREST


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model4=RandomForestClassifier()
model4.fit(xtrain,ytrain)
y_pred=model4.predict(xtest)
```

```python
acc4=accuracy_score(ytest,y_pred)
print(acc4)
```

```python
cm4=confusion_matrix(ytest,y_pred)
print(cm4)
dis4=ConfusionMatrixDisplay(confusion_matrix=cm4)
dis4.plot()
plt.show()
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
model5=LogisticRegression()
model5.fit(xtrain,ytrain)
y_pred=model5.predict(xtest)
```

```python
acc5=accuracy_score(ytest,y_pred)
print(acc5)
```

```python
cm5=confusion_matrix(ytest,y_pred)
print(cm5)
dis5=ConfusionMatrixDisplay(confusion_matrix=cm5)
dis5.plot()
plt.show()
```

comparison


```python
df=pd.DataFrame({'Accuracy':[acc1,acc2,acc3,acc4,acc5],'Algorithm':['KNN','NAIVE BAYES','DECISION TREE','RANDOM FOREST','LOGISTIC']})
df
```

```python
plt.figure(figsize=(10, 6))

plt.plot('Algorithm','Accuracy',data=df,marker='d',color='blue',ms = 10,mfc = 'r',mec = 'black')
plt.title("Accuracy Score")
plt.xlabel("Algorithms")
plt.ylabel("Accuracy")
plt.plot()
plt.grid()
```
